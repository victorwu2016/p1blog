Three questions regarding job security.
    1. How company size affects people to strong agree job security?
    2. How company type infulences people's feeling of job security?
    3. For the people who strong agree job security, how about their job satisfication and salaries?

Step 1: Prepare for data set

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
%matplotlib inline

df = pd.read_csv('./survey_results_public.csv')
df.head()
```

Step 2: Prepare data

```python
job_security_df = df[df.JobSecurity == "Strongly agree"]
```

Question 1: How company size affects people to strong agree job security?

```python
#Drop the rows with missing CompanyType values
company_size_df = job_security_df.dropna(subset=['CompanySize'], axis=0)
company_size_1 = company_size_df.CompanySize.value_counts()

(company_size_1/company_size_df.shape[0]).plot(kind="bar");
plt.xlabel("Company size")
plt.ylabel("Percentage")
plt.title("Strongly Agree for Job Security: Company Size")
```

Question 2: How company type infulences people's feeling of job security?

```python
#Drop the rows with missing CompanyType values
company_type_df = job_security_df.dropna(subset=['CompanyType'], axis=0)


company_type_1 = company_type_df.CompanyType.value_counts()

(company_type_1/company_type_df.shape[0]).plot(kind="bar");
plt.xlabel("Company type")
plt.ylabel("Percentage")
plt.title("Strongly Agree for Job Security: Company Type")
```

```python
def get_heatmap(df):
    '''
    FUNCTION:
    plots a seaborn heatmap based on the correlation
    
    INPUT:
    df - the pandas dataframe you want to present
    
    OUTPUT:
    matplotlib Axes
    '''
    return sns.heatmap(df,annot=True, fmt=".2f")
```

Question 3: For the people who strong agree job security, how about their job satisfication and salaries?
